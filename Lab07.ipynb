```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

**get the Data**
scikit learn

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
np.sum(pd.isnull(df).sum())
```

what are the 'target' data in the data set?

```python
cancer['target']
```

adding the target data to the dataframe

```python
df['cancer']=pd.DataFrame(cancer['target'])
df.head()
```

```python
sns.set_style('whitegrid')
```

```python
sns.boxplot(data=df.iloc[:, :10])
plt.title('Boxplot of First 10 Features')
plt.show()
```

**train test split**

```python
df_feat = df.drop('cancer',axis=1)
df_feat.head()
```

Prepare the Train and Test data from the dataframe.
Drop the cancer columns and define the dataframe with only target results

```python
df_target=df['cancer']
df_target.head()
```

```python
from sklearn.model_selection import train_test_split
```

Split the data into train, test using train test split

```python
X_train, X_test, y_train, y_test = train_test_split(df_feat, df_target, test_size=0.30, random_state=101)
```

```python
y_train.head()
```

**Train the SVC using the Train Dataset**

```python
from sklearn.svm import SVC
```

Apply the model.fit to dataset

```python
model=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
```

Predict and Analysis the Performance of the SVC Model
â€¢ Apply model.predict(X test)
â€¢ Generate the classification report, confusion matrix using sklearn.metrics

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,prediction))
```

```python
print(classification_report(y_test,prediction))
```

```python
print("misclassification error rate:",round(np.mean(prediction!=y_test),3))
```

Improve the Accuracy of Model using GridSearchCV Model

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 'kernel': ['rbf']}
grid = GridSearchCV(SVC(), param_grid, refit=True, verbose=3)
grid.fit(X_train, y_train)
```

```python
# Best parameters
print("Best Parameters:", grid.best_params_)
```

```python
# Predict using the best model
y_pred_grid = grid.predict(X_test)
```

```python
# Evaluate the performance
print("Classification Report with GridSearchCV")
classification_report(y_test, y_pred_grid)
```

```python
confusion_matrix(y_test, y_pred_grid)
```
